```python
import pandas as pd
```

```python
data = pd.read_csv("openbeerdb_csv/beers.csv", low_memory=False, usecols=["name", "abv"], nrows=100)
```

```python
data["abv"] = data["abv"].astype(float)
```

```python
data.plot(kind="barh", x="name", y="abv", figsize=(20, 20))
```

```python
scraper = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_beer_consumption_per_capita", attrs={"class": "wikitable"})
```

```python
table = scraper[0]
```

```python
table["Country"] = table["Country"].apply(lambda c: c.upper().replace("*", ""))
```

```python
table = table[["Country", "Consumptionper capita[1](litres per year)"]]
```

```python
table.plot(kind="barh", y=1, x=0, figsize=(20, 20))
```
